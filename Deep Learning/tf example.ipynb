## Pre-processing the Data

Changing the type of the datasets from a string value to an integer. This is so we can use the type as a label.

```python
# Converting the data into a numerical value
white['type'] = 0
red['type']   = 1
```

```python
# A new column has been added to the dataset
white.head()
```

We will connect the two datasets to be one, to feed into the ANN.

```python
# ignore_index=True so not duplicate index labels from joining both data sets together
wines = pd.concat([red, white], ignore_index=True)
wines
```

Creating values for the data (X) and labels y.

```python
X = wines # storing data against X variable 
y = np.ravel(wines.type) # gives us a 1D array and print Y variable to compare for later.
```

```python
print(y)
```

```python
# Dropping the targets from the dataframe
X = X.drop(columns=['type'])
```

## Creating training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X_train.shape)
print(X_test.shape)

"""
For now, import the train_test_split from sklearn.model_selection and 
assign the data and the target labels to the variables X and y

"""
```

### Standardising the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
""" 
Some of the values seen in the visualisations are far part. 
Standardization will scale the values to a normal range 0-1

""" 

scalar = StandardScaler().fit(X_train)
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
```

**Creating the ANN**

```python
import tensorflow as tf

from tensorflow              import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

creating the model:

```python
"""
Since neural networks can only work with numerical data, you have already encoded red as 1 and white as 0. 
(Sequential model is 1 input 1 output)
"""
model = Sequential()
```

creating the model's layers:

```python
# Input layer
model.add(Dense(12, activation='relu', input_shape=(12,)))
# which is the relu activation function (linear algorithm) transforms the numbers to 0 or the input itself


#Hidden layer
model.add(Dense(8, activation='relu'))
# Dense = fully connected

# Output label
model.add(Dense(1, activation='sigmoid'))
# sigmoid is a logistical algorithm and brings the results to a 0-1 
```

Exploring the models parameters:

```python
# Checking output shape
print("Output shape:", model.output_shape)
```

```python
# Viewing the randomly assigned weights
print("Model weights:", model.get_weights())
```

Compile the model:

```python
model.compile(loss='binary_crossentropy', optimizer='adam',
              metrics=['accuracy'])

"""
Loss calculates the percentage of what it got wrong. The binary_crossentropy is the default function 
for binary classification as we are looking to see if it is red or white wine. 
Optimizer then re-runs it, The choice (adam) is how the network learns (calculates)
metrics =results type, we looking for accuracy 

"""
```

Fit the model:

```python
model.fit(X_train, y_train, epochs=20,
          batch_size=1, verbose=1)

"""
Epochs, number of times it runs through 
Batch size is number of data sets
Verbose – how we want to see the data 

"""
```

Predict with the model:


```python
"""
Passing in test data into network. Model has not seen this before only seen the train data.
Calculates are in Scientific Notation and first 15 are printed out. 

"""
y_pred = model.predict(X_test)
y_pred[:15]
```

```python
"""
Converting into values between 0-1 and if statements used to determine if it is red or white wine. 
1= red 0= white

"""

predictions = [1 if p > 0.5 else 0 for p in y_pred]
```

```python
# Printing out first 15 predictions 
predictions[:15]
```

```python
# Y_test is the actual values so we can compare how well it did.
y_test[:15]
```

Evaluating the model:

```python
# Accuracy is how well the model did predict overall.
score = model.evaluate(X_test, y_test, verbose=1)
score
```

```python
from sklearn.metrics import confusion_matrix, precision_score
```

```python
# Breakdown of predictoions into a table
confusion_matrix(y_test, predictions)
```

```python
# Precision is measure og a classifier's exactness. The higher the number the more accurtae the classifer
precision_score(y_test, y_pred.round())
```
